# Perspective Projection Model

## Basic Principles

The topic to be discussed here is: how digital images are formed. Expanding this question into a more detailed one, that is, if the 3D coordinates of an object in a scene are given, how can we calculate the projection coordinates of this object in a 2D image?

Perspective projection is a method for calculating the position of 3D points in a 2D image. It takes into account the camera's intrinsic and extrinsic parameters, as well as the distance from the object to the camera. Through perspective projection, we can map points in the 3D coordinate system onto the 2D image plane.

The perspective projection model is shown in the figure below:

  <img src=".\\Practice_2_Imaging_System_Illustrations\\Figure 1_Perspective_Projection_Model.png" alt="Brief description" width="60%">
  <p>Figure 1 Perspective Projection Model</p>
</div>

The mathematical expression is written as follows:
  $$
    \lambda
    \begin{bmatrix}
    u \\
    v \\
    1
    \end{bmatrix}
    =
    \begin{bmatrix}
    f/du & 0 & u_0 \\
    0 & f/dv & v_0 \\
    0 & 0 & 1
    \end{bmatrix}
    \begin{bmatrix}
    R & T
    \end{bmatrix}
    \begin{bmatrix}
    X_w \\
    Y_w \\
    Z_w \\
    1
    \end{bmatrix}
  $$

Here, we can define

$$
K = 
\begin{bmatrix}
f/du & 0 & u_0 \\
0 & f/dv & v_0 \\
0 & 0 & 1
\end{bmatrix}
$$

In this context, matrix $K$ is the camera intrinsic matrix, which contains the camera's focal length and principal point coordinates. Meanwhile, $[R|T]$ is the camera extrinsic matrix, describing the camera's position and orientation in the world coordinate system.

Furthermore, considering the distortion inherent in the lens itself, the perspective projection model requires correction. Common distortion models include radial distortion and tangential distortion. Radial distortion causes objects in the image to appear compressed or stretched, while tangential distortion causes objects to appear rotated. Figure 2 illustrates the effect of radial distortion.

<div align="center">
  <img src=".\\Practice_2_Imaging_System_Illustrations\\Figure 2_Radial_Distortion.png" alt="Brief description" width="60%">
  <p>Figure 2 Schematic diagram of radial distortion</p>
</div>

To correct distortion in the perspective projection model, we need to introduce distortion coefficients. The distortion coefficient is a vector containing parameters for both radial and tangential distortion. By introducing these coefficients, we can map points from the perspective projection model onto an undistorted image plane. The method employed is essentially a simple polynomial fitting approach, as shown below:

$$
\begin{bmatrix}
u_d \\
v_d
\end{bmatrix}
= (1 + k_1 r^2)
\begin{bmatrix}
u - u_0 \\
v - v_0
\end{bmatrix}
+
\begin{bmatrix}
u_0 \\
v_0
\end{bmatrix}
$$

In this equation, $u_d$ and $v_d$ are the distorted image coordinates, $u$ and $v$ are the undistorted image coordinates, $u_0$ and $v_0$ are the principal point coordinates, $k_1$ is the radial distortion coefficient, and $r^2$ is the square of the distance from the point to the principal point, calculated as follows:

$$
r^2 = (u - u_0)^2 + (v - v_0)^2
$$

Thus, we have obtained a relatively complete perspective projection model. Through this model, we can map points from a three-dimensional coordinate system onto a two-dimensional image plane.
